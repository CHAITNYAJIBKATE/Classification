```python
#gender
ds_train.groupby('Gender').agg({'Gender': np.size})
ds_train["Gender"]=ds_train["Gender"].fillna('Male')
ds_train["Gender"] = ds_train["Gender"].astype('category')
ds_train["Gender"] = ds_train["Gender"].cat.codes
#married
ds_train.groupby('Married').agg({'Married': np.size})
ds_train["Married"]=ds_train["Married"].fillna('Yes')
ds_train["Married"] = ds_train["Married"].astype('category')
ds_train["Married"] = ds_train["Married"].cat.codes
#Dependents 
ds_train.groupby('Dependents').agg({'Dependents': np.size})
ds_train["Dependents"]=ds_train["Dependents"].fillna(0)
ds_train["Dependents"] = ds_train["Dependents"].astype('category')
ds_train["Dependents"] = ds_train["Dependents"].cat.codes
#Self_Employed 
ds_train.groupby('Self_Employed').agg({'Self_Employed': np.size})
ds_train["Self_Employed"]=ds_train["Self_Employed"].fillna('No')
ds_train["Self_Employed"] = ds_train["Self_Employed"].astype('category')
ds_train["Self_Employed"] = ds_train["Self_Employed"].cat.codes
```

```python
#LoanAmount
ds_train["LoanAmount"].fillna(ds_train["LoanAmount"].mean(), inplace=True)
#Loan_Amount_Term
ds_train["Loan_Amount_Term"].fillna(ds_train["Loan_Amount_Term"].mean(), inplace=True)
#Credit_History
ds_train["Credit_History"].fillna(ds_train["Credit_History"].mean(), inplace=True)
```
